```python
import cv2
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt

device = torch.device('cpu')

# Load the YOLOv8 model
model = YOLO("best.pt")

model.to(device)

# Load an input image
input_image_path = r"C:\Users\91630\Desktop\Final project\maxresdefault.jpg"
input_image = cv2.imread(input_image_path)

# Run YOLOv8 tracking on the input image
results = model.track(input_image, persist=True, verbose=False)  # Suppress performance metrics

# Visualize the results on the frame
annotated_frame = results[0].plot()

# Display the input and annotated images using matplotlib
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title("Input Image")
#plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(annotated_frame)
plt.title("Annotated Image")
#plt.axis('off')

plt.show()
```
